```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('/home/ospcx/GitProjects/gda_analyzer/data/unstructured/frecuencia_adultos_colapsada.csv', sep=',')
pantro = pd.read_csv('/home/ospcx/GitProjects/gda_analyzer/data/unstructured/pantro_adultos.csv', sep=',')
pantro =  pantro[['folio', 'peso', 'talla', 'imc']]
data.head() 
```

```python
desired_cols = [
    'folio',
    'sexo',
    'edaddias',
    'energ_kcal',
    'carbohydrt',
    'lipid_tot',
    'protein',
    'sugar_tot',
    'sodium',
    'fa_trans'
]
```

```python
subset_df = data[desired_cols]
subset_df.head(10)
```

```python
df_merged = pd.merge(subset_df, pantro, left_on='folio', right_on='folio')
df_merged.head()
```

```python
df_merged[['peso', 'talla', 'imc']] = df_merged[['peso', 'talla', 'imc']].astype(str)
# df_merged[['peso', 'talla', 'imc']] = df_merged[['peso', 'talla', 'imc']].replace(' ', 'nan', inplace=True)
df_merged = df_merged.sort_values(['peso', 'talla', 'imc'])
df_merged.to_csv('/home/ospcx/GitProjects/gda_analyzer/data/unstructured/clean.csv', sep=',', index=False)
```
